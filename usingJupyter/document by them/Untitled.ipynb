## Stack and Queue

```python
stack = []

# Push
stack.append(10)
stack.append(20)

# Pop
stack.pop()

# Peek
top = stack[-1]

# Is Empty
if not stack:
    print("Empty")
```

```python
class Stack:
    def __init__(self):
        self.stack = []

    def push(self, x):
        self.stack.append(x)

    def pop(self):
        if self.is_empty():
            return "Underflow"
        return self.stack.pop()

    def peek(self):
        return self.stack[-1]

    def is_empty(self):
        return len(self.stack) == 0
```

```python
s=Stack()
```

```python
s.push(3)
```

```python
s.stack
```

```python
s.pop()
```

```python
s.peek()
```

## Queue

A deque (double-ended queue), pronounced "deck," is a list-like data structure in Python's collections module that allows efficient O(1) appends and pops from both the left and right ends. This makes it a better choice than a standard Python list for implementing queues and stacks, as list operations at the beginning have O(n) time complexity. 

>we use deque because we can remove element in both side by pop() and popleft() and add by append() to right and to left we use appendleft()

```python
from collections import deque
p=deque()

for i in (0,5):
    p.append(i)

p.popleft()
```

```python
class Queue:
    def __init__(self):
        self.q = []

    def enqueue(self, x):
        self.q.append(x)

    def dequeue(self):
        if not self.q:
            return "Underflow"
        return self.q.pop(0)

    def front(self):
        return self.q[0]
```
